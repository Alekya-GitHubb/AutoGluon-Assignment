```python
from autogluon.tabular import TabularDataset, TabularPredictor

train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
test_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')

print("Train shape:", train_data.shape)
train_data.head()
```

```python
label = 'class'
save_path = 'ag_tabular_indepth/'   # folder to store models

predictor = TabularPredictor(
    label=label,
    path=save_path,
    problem_type='classification'
).fit(
    train_data,
    presets='best_quality',   # more thorough training
    time_limit=300            # 5 minutes
)
```

```python
label = 'class'
save_path = 'ag_tabular_indepth/'

predictor = TabularPredictor(
    label=label,
    path=save_path,
    problem_type='binary'  # <-- fixed here
).fit(
    train_data,
    presets='best_quality',
    time_limit=300
)
```

```python
leaderboard = predictor.leaderboard(train_data, silent=True)
leaderboard
```

```python
importance = predictor.feature_importance(train_data)
importance.head(10)
```

```python
performance = predictor.evaluate(test_data)
print(performance)
```
