```python
from autogluon.tabular import TabularDataset, TabularPredictor

# Load sample multimodal dataset (text + numeric + categorical)
train_data = TabularDataset("https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv")
test_data = TabularDataset("https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv")

train_data.head()
```

```python
label = 'class'
save_path = 'ag_tabular_multimodal/'

predictor = TabularPredictor(label=label, path=save_path, problem_type='binary').fit(
    train_data,
    presets='medium_quality_faster_train',  # Faster preset for mixed data
    time_limit=300
)
```

```python
predictor.leaderboard(train_data)
```

```python
predictor.feature_importance(train_data)
```
