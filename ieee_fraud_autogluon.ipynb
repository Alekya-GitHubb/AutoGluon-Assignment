```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

# ✅ Step 1: Load full dataset
train_data_full = pd.read_csv("ieee_fraud_data/train_transaction.csv")

# ✅ Step 2: Check dataset size
print("Full dataset shape:", train_data_full.shape)

# ✅ Step 3: Take only 10% of the data to avoid RAM crash
train_data = train_data_full.sample(frac=0.10, random_state=42)

print("Subset data shape:", train_data.shape)

# ✅ Step 4: Define label & save path
label = 'isFraud'
save_path = 'ag_fraud_model_10percent/'

# ✅ Step 5: Train AutoGluon model on subset
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)

# ✅ Step 6: View model leaderboard
leaderboard = predictor.leaderboard(train_data, silent=True)
leaderboard
```

```python
from autogluon.tabular import TabularPredictor

label = 'isFraud'
save_path = 'ag_fraud_model_10percent/'  # folder to store model

# Start training
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)

# Display leaderboard of models trained
leaderboard = predictor.leaderboard(train_data, silent=True)
leaderboard
```

```python
leaderboard = predictor.leaderboard(train_data, silent=True)
leaderboard
```

```python
# Make predictions
preds = predictor.predict(train_data)
preds.head()
```

```python
from autogluon.tabular import TabularPredictor

# Evaluate performance on same training subset (for demo)
performance = predictor.evaluate(train_data)
performance
```
