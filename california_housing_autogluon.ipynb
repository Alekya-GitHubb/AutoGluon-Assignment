```python
from autogluon import core
import platform
print("AutoGluon Version:", core.__version__)
print("Python Version:", platform.python_version())
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
data = fetch_california_housing(as_frame=True)
df = data.frame

# View sample
df.head()
print("Shape:", df.shape)
```

```python
from autogluon.tabular import TabularPredictor, TabularDataset

# Convert to autogluon format
train_data = TabularDataset(df)

label = 'MedHouseVal'   # Median house value (target)

predictor = TabularPredictor(label=label, problem_type='regression', path='ag_california_model/').fit(train_data)

# Show model leaderboard
predictor.leaderboard(train_data)
```
